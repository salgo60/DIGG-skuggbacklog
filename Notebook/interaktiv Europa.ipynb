```python
import folium

countries = {
    "Sweden": ("Hybrid", 60, 15),
    "Finland": ("Centralized", 64, 26),
    "Denmark": ("Centralized", 56, 10),
    "Norway": ("Hybrid", 62, 10),
    "Iceland": ("Hybrid", 65, -18),
    "Germany": ("Decentralized", 51, 10),
    "France": ("Centralized", 46, 2),
    "Netherlands": ("Hybrid", 52, 5),
    "Belgium": ("Decentralized", 50.5, 4.5),
    "Spain": ("Hybrid", 40, -4),
    "Portugal": ("Centralized", 39.5, -8),
    "Italy": ("Hybrid", 42.5, 13.5),
    "Austria": ("Hybrid", 47.5, 14),
    "Switzerland": ("Hybrid", 47, 8),
    "Czechia": ("Centralized", 49.8, 15.5),
    "Poland": ("Hybrid", 52, 19),
    "Estonia": ("Centralized", 58.6, 25),
    "Latvia": ("Centralized", 56.8, 25),
    "Lithuania": ("Centralized", 55.3, 24),
    "United Kingdom": ("Centralized", 54, -2),
    "Ireland": ("Centralized", 53, -8),
}
```

```python
colors = {
    "Centralized": "red",
    "Hybrid": "orange",
    "Decentralized": "blue"
}
```

```python
m = folium.Map(location=[54, 15], zoom_start=4)
```

```python
for country, (model, lat, lon) in countries.items():
    folium.CircleMarker(
        location=[lat, lon],
        radius=8,
        popup=f"{country}<br>Model: {model}",
        color=colors[model],
        fill=True,
        fill_color=colors[model]
    ).add_to(m)
```

```python
m
```

```python
from folium.plugins import MarkerCluster

cluster = MarkerCluster().add_to(m)

for country, (model, lat, lon) in countries.items():
    folium.Marker(
        [lat, lon],
        popup=f"<b>{country}</b><br>{model}",
        icon=folium.Icon(color="green")
    ).add_to(cluster)
```

```python
m
```

```python
legend_html = """
<style>
#legendBox {
    position: fixed;
    bottom: 20px;
    left: 20px;
    width: 160px;
    z-index: 9999;
    font-family: Arial, sans-serif;
}

#legendHeader {
    background: #333;
    color: white;
    padding: 6px;
    cursor: pointer;
    font-size: 14px;
    border-radius: 4px;
    text-align: center;
}

#legendContent {
    background: white;
    border: 2px solid #555;
    border-radius: 4px;
    padding: 10px;
    margin-top: 5px;
    display: none; /* Hidden by default */
    font-size: 13px;
}

.legend-color-box {
    width: 12px;
    height: 12px;
    display: inline-block;
    margin-right: 6px;
    vertical-align: middle;
}
</style>

<div id="legendBox">
  <div id="legendHeader">Legend </div>
  <div id="legendContent">
    <div><span class="legend-color-box" style="background:red;"></span>Centralized</div>
    <div><span class="legend-color-box" style="background:orange;"></span>Hybrid</div>
    <div><span class="legend-color-box" style="background:blue;"></span>Decentralized</div>
  </div>
</div>

<script>
document.getElementById('legendHeader').onclick = function() {
    var content = document.getElementById('legendContent');
    if (content.style.display === "none") {
        content.style.display = "block";
        this.innerHTML = "Legend ";
    } else {
        content.style.display = "none";
        this.innerHTML = "Legend ";
    }
};
</script>
"""

m.get_root().html.add_child(folium.Element(legend_html))
```
