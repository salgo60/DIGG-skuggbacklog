```python
import time

from datetime import datetime

now = datetime.now()
timestamp = now.timestamp()

start_time = time.time()
print("Start:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 5)

# 1. läs in
df = pd.read_json("forum_data.json")

# Kolla snabbt
df.head(), df.info()
```

```python
df.head(), df.info()
```

```python
#Hur många topics, inlägg, personer?  
n_topics = df["tid"].nunique()
n_posts  = len(df)
n_users  = df["username_norm"].nunique()

print("Antal topics:", n_topics)
print("Antal inlägg:", n_posts)
print("Antal användare:", n_users)
```

```python
#Mest aktiva personer
posts_per_user = (
    df.groupby("username_norm")["pid"]
      .count()
      .sort_values(ascending=False)
)

print(posts_per_user.head(20))

# enkel barplot topp 15
posts_per_user.head(15).plot(kind="bar")
plt.title("Topp 15 mest aktiva användare")
plt.ylabel("Antal inlägg")
plt.tight_layout()
plt.show()
```

```python
#  Inlägg över tid (hela forumet) 
posts_per_month = (
    df.groupby("post_month")["pid"]
      .count()
      .sort_index()
)

print(posts_per_month.head())

posts_per_month.plot(kind="line", marker="o")
plt.title("Antal inlägg per månad")
plt.xlabel("Månad")
plt.ylabel("Antal inlägg")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3) Klassificera inlägg (dataportalen, dataset, EDP, mm) 
from bs4 import BeautifulSoup

def html_to_text(html):
    if pd.isna(html):
        return ""
    return BeautifulSoup(html, "html.parser").get_text(separator=" ", strip=True)

df["content_text"] = df["content_html"].apply(html_to_text)
df["content_lower"] = df["content_text"].str.lower()
categories = {
    "dataportal_allmänt": ["dataportal.se", "sveriges dataportal", "community.dataportal"],
    "dataset_badplatser": ["badplats", "badvatten", "eubathingwateridentifier"],
    "dataset_utegym": ["utegym"],
    "dataset_grillplatser": ["grillplats", "grillplatser"],
    "edp_data_europa": ["european data portal", "data.europa.eu", "edp"],
    "termkataloger": ["termkatalog", "begrepp", "termer", "vokabulär"],
}
def classify_post(text):
    hits = []
    for cat, words in categories.items():
        if any(w in text for w in words):
            hits.append(cat)
    if not hits:
        return "oklassificerad"
    return ";".join(hits)

df["category"] = df["content_lower"].apply(classify_post)
print(df["category"].value_counts().head(20))

df["category"].value_counts().head(15).plot(kind="bar")
plt.title("Fördelning av kategorier (topp 15)")
plt.ylabel("Antal inlägg")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# 4.1) DIGG‑aktivitet (med och utan Maria) 
df_digg = df[df["is_digg"]]

posts_per_digg = (
    df_digg.groupby("username_norm")["pid"]
           .count()
           .sort_values(ascending=False)
)

print(posts_per_digg)

posts_per_digg.plot(kind="bar")
plt.title("Antal inlägg per DIGG‑användare")
plt.ylabel("Antal inlägg")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# 4.2. DIGG över tid – med vs utan Maria
# med alla DIGG
digg_month = (
    df_digg.groupby("post_month")["pid"]
           .count()
           .sort_index()
)

# exkludera Maria_Dalhage
df_digg_no_maria = df_digg[df_digg["username_norm"] != "maria_dalhage"]
digg_no_maria_month = (
    df_digg_no_maria.groupby("post_month")["pid"]
                    .count()
                    .sort_index()
)

# totala inlägg samma skala
total_month = (
    df.groupby("post_month")["pid"]
      .count()
      .sort_index()
)

# plott: total vs digg vs digg utan Maria
plt.plot(total_month.index, total_month.values, label="Alla")
plt.plot(digg_month.index, digg_month.values, label="DIGG (alla)")
plt.plot(digg_no_maria_month.index, digg_no_maria_month.values, label="DIGG (utan Maria)")
plt.legend()
plt.title("DIGG forum aktivitet - Inlägg per månad – totalt och DIGG")
plt.xlabel("Månad")
plt.ylabel("Antal inlägg")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# 4.3. Skapade trådar vs svar (DIGG) 
# Anta att post_index == 0 betyder första posten i en tråd
digg_topics_created = df_digg[df_digg["post_index"] == 0]
digg_replies         = df_digg[df_digg["post_index"] > 0]

print("DIGG – antal trådar de startat:", digg_topics_created["tid"].nunique())
print("DIGG – antal svar (poster med post_index>0):", len(digg_replies))

print("\nAndelar av alla trådar/inlägg:")
print("Andel trådar med DIGG som första post:",
      digg_topics_created["tid"].nunique() / df["tid"].nunique())

print("Andel inlägg som är DIGG‑inlägg:",
      len(df_digg) / len(df))
```

```python
# Utan Maria

digg_no_maria = df_digg[df_digg["username_norm"] != "maria_dalhage"]
digg_nm_topics_created = digg_no_maria[digg_no_maria["post_index"] == 0]
digg_nm_replies        = digg_no_maria[digg_no_maria["post_index"] > 0]
print("DIGG – antal trådar de startat utan Maria:", digg_nm_topics_created["tid"].nunique())
print("DIGG – antal svar utan Maria (poster med post_index>0):", len(digg_nm_replies))

print("\nAndelar av alla trådar/inlägg utan Maria:")
print("Andel trådar med DIGG som första post utan Maria:",
      digg_nm_topics_created["tid"].nunique() / df["tid"].nunique())

print("Andel inlägg som är DIGG‑inlägg utan Maria:",
      len(digg_no_maria) / len(digg_no_maria))
```

```python
 # End timer and calculate duration
end_time = time.time()
elapsed_time = end_time - start_time# Bygg audit-lager för den här etappen

# Print current date and total time
print("Date:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
minutes, seconds = divmod(elapsed_time, 60)
print("Total time elapsed: {:02.0f} minutes {:05.2f} seconds".format(minutes, seconds))
```
