```python
import time

from datetime import datetime

now = datetime.now()
timestamp = now.timestamp()

start_time = time.time()
print("Start:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
```

```python
import json
import pandas as pd

with open("regb_master_poc.json", "r", encoding="utf-8") as f:
    regb = json.load(f)

rows = []

for year_obj in regb["years"]:
    year = year_obj["year"]
    for task in year_obj["tasks"]:
        rows.append({
            "year": year,
            "task_id": task["id"],
            "title": task["title"],
            "eu_relevance": task["eu_relevance"],
            "metadata_relevance": task["metadata_relevance"],
            "manual_regex_count": len(task["forum_regex"]["manual"]),
            "ml_regex_count": len(task["forum_regex"]["ml"])
        })

df = pd.DataFrame(rows)
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

tasks_per_year = df.groupby("year").size()

plt.figure(figsize=(10,4))
tasks_per_year.plot(kind="bar", color="steelblue")
plt.title("Antal uppdrag per år (DIGG regleringsbrev)")
plt.ylabel("Antal uppdrag")
plt.xlabel("År")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
meta_per_year = df[df["metadata_relevance"] == True].groupby("year").size()

plt.figure(figsize=(10,4))
meta_per_year.plot(marker="o", linewidth=2, color="darkgreen")
plt.title("Tidslinje: Metadata‑relevanta uppdrag per år")
plt.ylabel("Antal metadata‑uppdrag")
plt.xlabel("År")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
eu_per_year = df[df["eu_relevance"] == True].groupby("year").size()

plt.figure(figsize=(10,4))
eu_per_year.plot(marker="o", linewidth=2, color="navy")
plt.title("Tidslinje: EU‑relevanta uppdrag per år")
plt.ylabel("Antal EU‑uppdrag")
plt.xlabel("År")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
## 🔥 5. Heatmap: metadata × EU × år 
import seaborn as sns
import matplotlib.pyplot as plt

pivot = df.groupby(["year", "metadata_relevance", "eu_relevance"]).size().reset_index()
pivot.columns = ["year", "metadata", "eu", "count"]

heat = pivot.pivot_table(index="year", columns=["metadata", "eu"], values="count", fill_value=0)

plt.figure(figsize=(10,6))
sns.heatmap(heat, annot=True, cmap="YlGnBu")
plt.title("Heatmap: metadata × EU × år")
plt.show()
```

```python
# 🧠 6. Bar chart: Regex‑komplexitet per år  
regex_complexity = df.groupby("year")["manual_regex_count"].sum()

plt.figure(figsize=(10,4))
regex_complexity.plot(kind="bar", color="orange")
plt.title("Regex‑komplexitet per år (manuella termer)")
plt.ylabel("Antal regex-termer")
plt.xlabel("År")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
 # End timer and calculate duration
end_time = time.time()
elapsed_time = end_time - start_time# Bygg audit-lager för den här etappen

# Print current date and total time
print("Date:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
minutes, seconds = divmod(elapsed_time, 60)
print("Total time elapsed: {:02.0f} minutes {:05.2f} seconds".format(minutes, seconds))
```
