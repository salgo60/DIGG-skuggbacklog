```python
import time

from datetime import datetime

now = datetime.now()
timestamp = now.timestamp()

start_time = time.time()
print("Start:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 5)

# 1. läs in
df = pd.read_json("forum_data.json")

# Kolla snabbt
df.head(), df.info()
```

```python
total_pm = df.groupby("post_month")["pid"].count()
digg_pm = df[df["is_digg"]].groupby("post_month")["pid"].count()

share = (digg_pm / total_pm * 100).fillna(0)

share.plot(marker="o")
plt.title("DIGG:s andel av inlägg över tid (%)")
plt.ylabel("Andel (%)")
plt.xlabel("Månad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("figures/08_digg_andel_over_tid.png", dpi=200)
plt.show()

plt.close()

```

```python
import numpy as np

posts_per_user = (
    df.groupby("username_norm")["pid"]
      .count()
      .sort_values()
)

cum_users = posts_per_user.cumsum() / posts_per_user.sum()

x = np.arange(len(cum_users))

plt.plot(x, cum_users.values, label="Faktisk fördelning")
plt.plot(x, x / x[-1], linestyle="--", label="Perfekt jämlikhet")

plt.title("Koncentration av aktivitet (Lorenz-kurva)")
plt.xlabel("Användare (sorterade)")
plt.ylabel("Ackumulerad andel inlägg")
plt.legend()
plt.tight_layout()
plt.savefig("figures/09_lorenz_koncentration.png", dpi=200)
plt.show()
plt.close()
```

```python
replies_per_topic = (
    df.groupby(["post_month", "tid"])["pid"]
      .count()
      .reset_index(name="posts")
)

replies_per_topic["replies"] = replies_per_topic["posts"] - 1

median_replies = (
    replies_per_topic
    .groupby("post_month")["replies"]
    .median()
)

median_replies.plot(marker="o")
plt.title("Median antal svar per tråd och månad")
plt.ylabel("Antal svar")
plt.xlabel("Månad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("figures/10_median_svar_per_trad.png", dpi=200)
plt.show()
plt.close()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# --------------------------------
# 1. Förbered data
# --------------------------------

# Identifiera trådstartare
thread_starters = (
    df[df["post_index"] == 0]
    .set_index("tid")["username_norm"]
)

# Ta bara svar
replies = df[df["post_index"] > 0].copy()

# Koppla svar -> trådstartare
replies["replied_to"] = replies["tid"].map(thread_starters)

# Ta bort själv-svar och NaN
edges = replies[
    (replies["username_norm"] != replies["replied_to"]) &
    (replies["replied_to"].notna())
][["username_norm", "replied_to"]]

# Räkna hur ofta A svarar B
edge_counts = (
    edges.groupby(["username_norm", "replied_to"])
          .size()
          .reset_index(name="weight")
)

# --------------------------------
# 2. Bygg graf
# --------------------------------
G = nx.DiGraph()

for _, row in edge_counts.iterrows():
    G.add_edge(
        row["username_norm"],
        row["replied_to"],
        weight=row["weight"]
    )

# --------------------------------
# 3. Rita grafen (enkel & pedagogisk)
# --------------------------------
plt.figure(figsize=(12, 12))

pos = nx.spring_layout(G, k=0.7, seed=42)

# Nodstorlek = hur aktiv totalt
node_sizes = [
    df[df["username_norm"] == n].shape[0] * 5
    for n in G.nodes()
]

# Kantbredd = antal svar
edge_widths = [
    G[u][v]["weight"] * 0.3
    for u, v in G.edges()
]

nx.draw_networkx_nodes(
    G, pos,
    node_size=node_sizes,
    node_color="lightblue",
    alpha=0.8
)

nx.draw_networkx_edges(
    G, pos,
    width=edge_widths,
    arrowstyle="->",
    arrowsize=12,
    alpha=0.6
)

nx.draw_networkx_labels(
    G, pos,
    font_size=8
)

plt.title("Vem svarar vem i forumet (aggregerad)")
plt.axis("off")
plt.tight_layout()
plt.savefig("figures/13_vem_svarar_vem.png", dpi=200)
plt.show()  
plt.close()

print("✅ Graf sparad som figures/13_vem_svarar_vem.png")
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# --------------------------------
# 1. Begränsa till topp 20 aktiva
# --------------------------------

user_activity = (
    df.groupby("username_norm")["pid"]
      .count()
      .sort_values(ascending=False)
)

top_users = (
    user_activity[user_activity >= 3]
    .head(20)
    .index
)

df_filt = df[df["username_norm"].isin(top_users)]

# --------------------------------
# 2. Identifiera trådstartare
# --------------------------------
thread_starters = (
    df_filt[df_filt["post_index"] == 0]
    .set_index("tid")["username_norm"]
)

# --------------------------------
# 3. Koppla svar -> trådstartare
# --------------------------------
replies = df_filt[df_filt["post_index"] > 0].copy()
replies["replied_to"] = replies["tid"].map(thread_starters)

edges = replies[
    (replies["username_norm"] != replies["replied_to"]) &
    (replies["replied_to"].notna())
][["username_norm", "replied_to"]]

edge_counts = (
    edges.groupby(["username_norm", "replied_to"])
          .size()
          .reset_index(name="weight")
)

# --------------------------------
# 4. Bygg graf
# --------------------------------
G = nx.DiGraph()

for _, row in edge_counts.iterrows():
    G.add_edge(
        row["username_norm"],
        row["replied_to"],
        weight=row["weight"]
    )

# --------------------------------
# 5. Rita graf
# --------------------------------
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.7, seed=42)

node_sizes = [
    df_filt[df_filt["username_norm"] == n].shape[0] * 20
    for n in G.nodes()
]

edge_widths = [
    G[u][v]["weight"] * 0.4
    for u, v in G.edges()
]

nx.draw_networkx_nodes(
    G, pos,
    node_size=node_sizes,
    node_color="lightblue",
    alpha=0.8
)

nx.draw_networkx_edges(
    G, pos,
    width=edge_widths,
    arrowsize=12,
    arrowstyle="->",
    alpha=0.6
)

nx.draw_networkx_labels(
    G, pos,
    font_size=9
)

plt.title("Vem svarar vem – topp 20 mest aktiva användare")
plt.axis("off")
plt.tight_layout()
plt.savefig("figures/13_vem_svarar_vem_top20.png", dpi=200)
plt.show()
plt.close()

print("✅ Graf sparad: figures/13_vem_svarar_vem_top20.png")
```

```python
# --------------------------------
# Burst-index per användare
# --------------------------------

user_month = (
    df.groupby(["username_norm", "post_month"])["pid"]
      .count()
      .reset_index(name="n_posts")
)

burst = (
    user_month.groupby("username_norm")
    .agg(
        total_posts=("n_posts", "sum"),
        max_month_posts=("n_posts", "max")
    )
)

burst["burst_index"] = burst["max_month_posts"] / burst["total_posts"]

# filtrera bort små konton
burst_filt = burst[burst["total_posts"] >= 5]

# topp 20 mest aktiva
top20 = (
    burst_filt
    .sort_values("total_posts", ascending=False)
    .head(20)
)

# --------------------------------
# Plot
# --------------------------------
top20.sort_values("burst_index").plot(
    y="burst_index",
    kind="barh",
    legend=False,
    color="steelblue"
)

plt.title("Kortvarig hög aktivitet per användare (Burst-index)")
plt.xlabel("Andel inlägg i mest aktiva månaden")
plt.tight_layout()
plt.savefig("figures/14_burst_index_top20.png", dpi=200)
plt.show()
plt.close()

print("✅ Sparade figures/14_burst_index_top20.png")
```

```python
df["post_time"] = pd.to_datetime(df["post_timestamp"])

user_stats = (
    df.groupby("username_norm")
    .agg(
        total_posts=("pid", "count"),
        first_post=("post_time", "min"),
        last_post=("post_time", "max")
    )
)

user_stats["active_days"] = (
    user_stats["last_post"] - user_stats["first_post"]
).dt.days

user_stats = user_stats[user_stats["total_posts"] >= 5]

plt.scatter(
    user_stats["active_days"],
    user_stats["total_posts"],
    alpha=0.6
)

plt.xlabel("Aktiv livslängd (dagar)")
plt.ylabel("Antal inlägg")
plt.title("Användarmönster: kortvarig vs långsiktig aktivitet")
plt.tight_layout()
plt.savefig("figures/15_livslangd_vs_volym.png", dpi=200)
plt.show()
plt.close()

print("✅ Sparade figures/15_livslangd_vs_volym.png")
```

```python
user_stats["active_months"] = (
    (user_stats["last_post"] - user_stats["first_post"]).dt.days / 30
)

user_stats["active_months"].plot.hist(bins=20)

plt.title("Hur länge användare är aktiva i forumet")
plt.xlabel("Aktiva månader")
plt.ylabel("Antal användare")
plt.tight_layout()
plt.savefig("figures/16_aktiva_manader.png", dpi=200)
plt.show()
plt.close()
```
