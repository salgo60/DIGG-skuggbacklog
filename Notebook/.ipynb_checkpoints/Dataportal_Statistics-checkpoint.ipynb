```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
import matplotlib.pyplot as plt

ENDPOINT = "https://admin.dataportal.se/sparql"

def run_query(q):
    sparql = SPARQLWrapper(ENDPOINT)
    sparql.setQuery(q)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    df = pd.DataFrame([{k:v["value"] for k,v in b.items()}
                       for b in results["results"]["bindings"]])
    # Rensa bort GraphDB:s interna namn
    df.columns = [c.replace("callret-0", "count") for c in df.columns]
    return df
```

```python
def to_numeric(df, col):
    df[col] = pd.to_numeric(df[col], errors="coerce")
    return df
```

### Step 2 — Find out what properties exist 
We’ll ask which RDF predicates (properties) are most common in datasets.

```python
query_props = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>

SELECT ?property (COUNT(?property) AS ?count)
WHERE {
  ?s a dcat:Dataset ;
     ?property ?o .
}
GROUP BY ?property
ORDER BY DESC(?count)
LIMIT 30
"""
df_props = run_query(query_props)
df_props
```

## Step 3 — Count total datasets

```python
query_count = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>
SELECT (COUNT(?dataset) AS ?total)
WHERE { ?dataset a dcat:Dataset . }
"""
run_query(query_count)
```

## Step 4 — See who delivers data (publishers)

```python
query_publishers = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>
PREFIX dct:  <http://purl.org/dc/terms/>
PREFIX foaf: <http://xmlns.com/foaf/0.1/>

SELECT ?publisher (COUNT(?dataset) AS ?datasets)
WHERE {
  ?dataset a dcat:Dataset ;
           dct:publisher ?org .
  ?org foaf:name ?publisher .
}
GROUP BY ?publisher
ORDER BY DESC(?datasets)
LIMIT 20
"""
df_pub = run_query(query_publishers)
df_pub["datasets"] = df_pub["datasets"].astype(int)
df_pub.plot(kind="barh", x="publisher", y="datasets",
            figsize=(8,6), color="skyblue")
plt.title("Topp 20 datapublicerare på Dataportal.se")
plt.xlabel("Antal dataset")
plt.gca().invert_yaxis()
plt.show()
```

## Step 5 — Graph: dataset growth over time

```python
query_growth = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>
PREFIX dct:  <http://purl.org/dc/terms/>

SELECT (year(?issued) AS ?year) (COUNT(?dataset) AS ?count)
WHERE {
  ?dataset a dcat:Dataset ;
           dct:issued ?issued .
  FILTER(BOUND(?issued))
}
GROUP BY (year(?issued))
ORDER BY ?year
"""
df_growth = run_query(query_growth)

# konvertera till rätt datatyper och filtrera bort tomma år
df_growth = df_growth[df_growth["year"].str.len() > 0]
df_growth["year"] = pd.to_numeric(df_growth["year"], errors="coerce")
df_growth["count"] = pd.to_numeric(df_growth["count"], errors="coerce")

df_growth = df_growth.dropna().sort_values("year")

plt.figure(figsize=(8,5))
plt.plot(df_growth["year"], df_growth["count"], marker="o", color="orange")
plt.title("Tillväxt av dataset på Dataportal.se (enligt dct:issued)")
plt.xlabel("År")
plt.ylabel("Antal dataset")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

## Step 7 - License usage statistics (dct:license)


```python
query_license = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>
PREFIX dct:  <http://purl.org/dc/terms/>

SELECT ?license (COUNT(?dataset) AS ?count)
WHERE {
  ?dataset a dcat:Dataset ;
           dct:license ?license .
}
GROUP BY ?license
ORDER BY DESC(?count)
LIMIT 20
"""
df_license = run_query(query_license)
# Hantera olika möjliga kolumnnamn för COUNT
count_col = "count" if "count" in df_license.columns else "callret-0"

df_license[count_col] = pd.to_numeric(df_license[count_col], errors="coerce")
df_license = df_license.dropna(subset=[count_col])
df_license = df_license.sort_values(count_col, ascending=False)

plt.figure(figsize=(8,5))
plt.barh(df_license["license"], df_license[count_col], color="lightgreen")
plt.title("De vanligaste licenserna på Dataportal.se")
plt.xlabel("Antal dataset")
plt.gca().invert_yaxis()
plt.show()


df_license
```
