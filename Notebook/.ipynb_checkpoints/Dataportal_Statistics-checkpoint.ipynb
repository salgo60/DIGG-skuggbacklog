```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
import matplotlib.pyplot as plt

ENDPOINT = "https://admin.dataportal.se/sparql"

def run_query(q):
    sparql = SPARQLWrapper(ENDPOINT)
    sparql.setQuery(q)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    df = pd.DataFrame([{k:v["value"] for k,v in b.items()}
                       for b in results["results"]["bindings"]])
    # Rensa bort GraphDB:s interna namn
    df.columns = [c.replace("callret-0", "count") for c in df.columns]
    return df
```

```python
def to_numeric(df, col):
    df[col] = pd.to_numeric(df[col], errors="coerce")
    return df
```

### Step 2 — Find out what properties exist 
We’ll ask which RDF predicates (properties) are most common in datasets.

```python
query_props = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>

SELECT ?property (COUNT(?property) AS ?count)
WHERE {
  ?s a dcat:Dataset ;
     ?property ?o .
}
GROUP BY ?property
ORDER BY DESC(?count)
"""
df_props = run_query(query_props)
df_props
```

```python
df_props.to_csv("props.csv")
```

## Step 3 — Count total datasets

```python
query_count = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>
SELECT (COUNT(?dataset) AS ?total)
WHERE { ?dataset a dcat:Dataset . }
"""
run_query(query_count)
```

## Step 4 — See who delivers data (publishers)

```python
query_publishers = """
PREFIX dcat: <http://www.w3.org/ns/dcat#>
PREFIX dct:  <http://purl.org/dc/terms/>
PREFIX foaf: <http://xmlns.com/foaf/0.1/>

SELECT ?publisher (COUNT(?dataset) AS ?datasets)
WHERE {
  ?dataset a dcat:Dataset ;
           dct:publisher ?org .
  ?org foaf:name ?publisher .
}
GROUP BY ?publisher
ORDER BY DESC(?datasets)
LIMIT 20
"""
df_pub = run_query(query_publishers)
df_pub["datasets"] = df_pub["datasets"].astype(int)
df_pub.plot(kind="barh", x="publisher", y="datasets",
            figsize=(8,6), color="skyblue")
plt.title("Topp 20 datapublicerare på Dataportal.se")
plt.xlabel("Antal dataset")
plt.gca().invert_yaxis()
plt.show()
```
