```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 5)

# 1. läs in
df = pd.read_json("forum_data.json")

# Kolla snabbt
df.head(), df.info()
```

```python
import requests
import time
from requests.exceptions import RequestException

BASE_URL = "https://community.dataportal.se/topic/"
SAVE_API = "https://web.archive.org/save/"
HEADERS = {"User-Agent": "ArchiveBot/1.0 (+your_email@example.com)"}

session = requests.Session()
session.headers.update(HEADERS)


def url_exists(url, retries=3):
    """Return True only if the source URL returns 200."""
    for attempt in range(1, retries + 1):
        try:
            r = session.head(url, timeout=20, allow_redirects=True)
            if r.status_code == 200:
                return True
            elif r.status_code >= 400:
                print(f"  ⚠ Source returned {r.status_code} → skip")
                return False
        except RequestException as e:
            print(f"  ⚠ HEAD error: {e} (attempt {attempt})")
            time.sleep(2 * attempt)

    print("  ❌ Could not verify URL → skip")
    return False


def archive_url(url, max_retries=5):
    """Try archiving a URL with retries and exponential backoff."""
    delay = 20

    for attempt in range(1, max_retries + 1):
        try:
            print(f"  Attempt {attempt}: {url}")

            r = session.get(
                SAVE_API + url,
                timeout=60,
                allow_redirects=True
            )

            if r.status_code in (200, 301, 302):
                saved = r.headers.get("Content-Location")
                if saved:
                    print("    ✔ Archived at:", "https://web.archive.org" + saved)
                else:
                    print("    ✔ Archived (no location returned)")
                return True

            print(f"    ✖ Status {r.status_code}, retrying...")

        except RequestException as e:
            print(f"    ✖ Error: {e}, retrying...")

        time.sleep(delay)
        delay *= 2

    print("    ❌ Failed after all retries:", url)
    return False


# ---- MAIN LOOP ----
df_sorted = df.sort_values("pid", ascending=False)

for row in df_sorted.itertuples(index=False):
    pid = row.pid
    url = f"{BASE_URL}{pid}"

    print(f"Checking {url} …")
    if not url_exists(url):
        continue

    print(f"Saving {url} …")
    archive_url(url)

    time.sleep(5)
```

```python
 # End timer and calculate duration
end_time = time.time()
elapsed_time = end_time - start_time# Bygg audit-lager för den här etappen

# Print current date and total time
print("Date:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
minutes, seconds = divmod(elapsed_time, 60)
print("Total time elapsed: {:02.0f} minutes {:05.2f} seconds".format(minutes, seconds))
```
